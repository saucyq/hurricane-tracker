```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python

dfDevastatorAt = pd.read_csv("data/"+"AL.csv")
dfDevastatorEP = pd.read_csv("data/"+"EP.csv")

```

```python
world = gpd.read_file("data/world.shp")
world.plot()
```

```python
hurricane = dfDevastatorAt[dfDevastatorAt["Status"]=="HU"]
colors = plt.cm.viridis(np.linspace(0, 1, len(range(1860, 2000, 40))))  

print(colors)

ax = world.plot(color="white", edgecolor="black", figsize=(20, 12))  # Plot the base map

for idx,i in enumerate(range(1860,2000,40)):
    hurricane= hurricane[(hurricane['DateTime'] >= str(i)) & (hurricane['DateTime'] <= str(i+40))]
    gdf = gpd.GeoDataFrame(
        hurricane, geometry=gpd.points_from_xy(hurricane.Lon, hurricane.Lat), crs="EPSG:4326"
    )

# We can now plot our ``GeoDataFrame``.
    gdf.plot(ax=ax,markersize=1, color=colors[idx])
plt.show()
```

```python
# We restrict to South America.
ax = world.plot(color="white", edgecolor="black",figsize=(20, 12))

# We can now plot our ``GeoDataFrame``.
gdf.plot(ax=ax,markersize=1, color="red")

#plt.show()
```
